## Cross Validation

* It help us to validate different ML models such as Linear regression, K-nearest neighbours, etc., and conclude which works best.
* For each method, it splits the data for **Testing** and **Training** and keep track of the output. Finally it tells us the efficiency of each method from which we can decide the most efficient method. 

**Types:**
* *Four fold cross validation*: Dividing the data into four sets.
* *Leave one out cross validation*: Leaving only one data element and using rest for the training.
